## Converting DICOM to JPEG

#### Importing Libraries

```python
import dicom
import scipy.misc
import os
import pandas as pd
import csv
```

#### Fetching current path

** Note: DICOM images are taken from hard disk, connect HD before running this script **

```python
cur_path = os.path.dirname(os.getcwd())
print(cur_path)
file_path = cur_path + '/data/calc_case_description_train_set.csv'
default_path = '/Volumes/easystore/cancer/CalcImg/DOI/'
```

#### Reading CSV file into Dataframe

```python
case_frame = pd.read_csv(file_path, encoding='latin-1')
```

```python
case_frame['ImageFilePath'] = ''
case_frame['ImageFilePath'] = default_path + case_frame['image file path']
```

```python
case_frame.head(1)
```

```python
case_frame['ImagePath'] = ''
```

#### Module to convert DICOME to JPEG format - lossless conversion

```python
def dicom_jpeg(case_frame):
    for index in range(1206,len(case_frame['ImageFilePath'])):
        ds = dicom.read_file(case_frame['ImageFilePath'][index])
        print(case_frame['ImageFilePath'][index])
        filename = cur_path + '/outcome/Calc_Data/' + str(index) + '.jpg'
        #case_frame['ImagePath'] = filename
        scipy.misc.imsave(filename, ds.pixel_array)
```

```python
#calling the function
dicom_jpeg(case_frame)
```

```python
case_frame.to_csv(cur_path +'/data/calc_case_description_train_set.csv', header=True,mode='w')
```
